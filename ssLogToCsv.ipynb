```python
import re
import csv

# === CONFIG ===
log_path = r"C:\Users\User\Desktop\tcp_results\Cake\20250703_170607_cubic_vs_bbrv3_upload\bbrv3\ss_logs\bbrv3_5202.log"
output_csv = r"C:\Users\User\Documents\GitHub\LCN_Demo\BBRv3_Features_Cake_Upload.csv"

# === OUTPUT HEADERS ===
headers = ['time', 'pacing_rate', 'delivery_rate', 'pacing_gain', 'inflight_data']
rows = []

with open(log_path, 'r') as f:
    time_marker = None
    pacing_rate_count = 0  # Count pacing_rate lines per time block

    for line in f:
        line = line.strip()

        # Extract time marker
        if line.startswith("==== Time"):
            match = re.search(r"Time\s+(\d+)", line)
            if match:
                time_marker = match.group(1)
                pacing_rate_count = 0  # Reset counter for new time block

        # Process only lines containing pacing_rate
        elif "pacing_rate" in line:
            pacing_rate_count += 1
            # Skip first pacing_rate per time block, take second only
            if pacing_rate_count == 2:
                pacing_rate = re.search(r"pacing_rate\s+(\d+[KMG]?bps)", line)
                delivery_rate = re.search(r"delivery_rate\s+(\d+[KMG]?bps)", line)
                pacing_gain = re.search(r"pacing_gain[:=]?([\d.]+)", line)
                inflight_data = re.search(r"inflight[:=]?(\d+)", line)

                row = [
                    time_marker,
                    pacing_rate.group(1) if pacing_rate else None,
                    delivery_rate.group(1) if delivery_rate else None,
                    pacing_gain.group(1) if pacing_gain else None,
                    inflight_data.group(1) if inflight_data else None,
                ]

                rows.append(row)

# === SAVE TO CSV ===
with open(output_csv, 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(headers)
    writer.writerows(rows)

print(f"\n✅ Saved to: {output_csv}")
```

```python
import re
import csv

# === CONFIG ===
log_path = r"C:\Users\User\Desktop\tcp_results\Cake\20250703_170607_cubic_vs_bbrv3_upload\bbrv3\ss_logs\bbrv3_5202.log"
output_csv = r"C:\Users\User\Documents\GitHub\LCN_Demo\BBRv3_Features_Cake_Upload.csv"

# === OUTPUT HEADERS ===
headers = ['time', 'pacing_rate', 'delivery_rate', 'pacing_gain', 'inflight_data']
rows = []

# === PARSE LOG ===
with open(log_path, 'r') as f:
    time_marker = None
    first_pacing_rate = False
    for line in f:
        line = line.strip()

        # Extract time marker
        if line.startswith("==== Time"):
            match = re.search(r"Time\s+(\d+)", line)
            if match:
                time_marker = match.group(1)
            print("line: ",line)
            print("match: ",match)
            print("time_marker: ",time_marker)
            first_pacing_rate = True



        # Process only lines containing pacing_rate
        elif "pacing_rate" in line and not(first_pacing_rate):

            pacing_rate = re.search(r"pacing_rate\s+(\d+[KMG]?bps)", line)
            delivery_rate = re.search(r"delivery_rate\s+(\d+[KMG]?bps)", line)
            pacing_gain = re.search(r"pacing_gain[:=]?([\d.]+)", line)
            inflight_data = re.search(r"inflight[:=]?(\d+)", line)

            print("------------------------------------------")

            print("pacing_rate_all :",pacing_rate)
            print("pacing_rate.group(1) :",pacing_rate.group(1))

            print("------------------------------------------")
            

            row = [
                time_marker,
                pacing_rate.group(1) if pacing_rate else None,
                delivery_rate.group(1) if delivery_rate else None,
                pacing_gain.group(1) if pacing_gain else None,
                inflight_data.group(1) if inflight_data else None,
            ]

            rows.append(row)
        
        else:
            first_pacing_rate = False

# === SAVE TO CSV ===
with open(output_csv, 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(headers)
    writer.writerows(rows)

print(f"\n✅ Saved to: {output_csv}")
```
